```python
import pandas as pd
import json

df = pd.read_csv('yelp.csv')

df.head()
```

```python
# summary statistics
summary_stats = df.describe()

summary_stats
```

```python
# evaluating null values
null_counts = df.isnull().sum()

null_columns = null_counts[null_counts > 0]
print(null_columns)
```

```python
# datatypes
df.dtypes
```

```python
# inspecting various attribute types
import ast

df['attributes'] = df['attributes'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

unique_keys = set()
for row in df['attributes'].dropna():
    if isinstance(row, dict):
        unique_keys.update(row.keys())

print(unique_keys)
```

```python
# 
all_categories = set()
df['categories'].dropna().apply(lambda x: all_categories.update(x.split(', ')))
print(all_categories)
```

```python
# drop unnecessary columns
df = df[['business_id', 'name', 'address', 'city', 'state', 'stars', 'review_count', 'is_open', 'attributes', 'categories']]
```

```python
df['is_open'] = df['is_open'].astype(bool)
```

```python
df['attributes'] = df['attributes'].apply(json.dumps)

print(df['attributes'])
```

```python
# filtering categories to ensure we are getting food or restaurant specific locations
df = df[df['categories'].str.contains(r'\b(Restaurants)\b', case=False, na=False)]

df = df.reset_index(drop=True)

df
```

```python
# downloading data
df.to_csv('cleaned_restaurant_data.csv', index=False)
```
